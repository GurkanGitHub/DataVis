# side by side bar plot with matplotlib

```python
labels = ["G1", "G2", "G3", "G4", "G5"]
x = np.arange(len(label)) 
```

```python
width = 0.35
```

```python
x - width/2
```

```python
x + width/2
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

# side by side bar plot with pandas

```python
label = ["G1", "G2", "G3", "G4", "G5"]
men_means = [20,34,30,35,27]
women_means=[25,32,34,20,25]
df = pd.DataFrame(list(zip(label, men_means, women_means)), columns = ["label", "men_means", "women_means"])
```

```python
df
```

```python
df.plot.bar()
```

```python
ax = df.plot.bar(rot=0, figsize = (8,4))
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
# ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.axhline(y=20, color="green", ls="--")


for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.05, p.get_height()+1))
```
